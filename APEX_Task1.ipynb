```python
df['Calculated_Sales'] = round((df['Quantity'] * df['Unit_Price']), 2)
```

```python
df['Difference'] = round((df['Total_Sales']- df['Calculated_Sales']), 2)
```

```python
(df['Difference'] != 0).sum()
```

```python
numeric_cols = ['Age', 'Quantity', 'Unit_Price', 'Total_Sales']

for col in numeric_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    print(f"{col}: {len(outliers)} outliers")
```

```python
# Till now i have done Data analyzation process now moving to the data clenaing process
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age'].isnull().sum()
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
```

```python
df['City'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
df['Quarter'] = df['Order_Date'].dt.quarter
```

```python
df['Year'] = df['Order_Date'].dt.year
```

```python
bins = [17,30,45,60,70]

labels = [
    'Young Adult',
    'Adult',
    'Middle Age',
    'Senior'
]

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels
)
```
